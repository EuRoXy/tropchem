# Demo of the influence of PBL mixing on the diurnal cycle of a trace gas
The routine pbl_evolution computes the diurnal evolution of an air pollution tracer in the planetary boundary layer for the following situation:
* the tracer is released at the surface with constant rate
* the tracer can be passive or have a fixed lifetime (in hours)
* a stable nocturnal inversion layer (0-200 m) is present from 21 local time (LT) in the evening to 6 LT in the next morning
* above this stable layer is a residual layer (200 m - 2000 m)
* each layer is internally well-mixed but there is no mixing between the layers
* exchange between the layers only occurs in the morning when the surface layer grows gradually incorporating air from the residual layer

## Examples:
* Evolution over 1 day, infinite lifetime: pbl_evolution()
* Evolution over 3 days, infinite lifetime: pbl_evolution(nday=3)
* Evolution over 3 days, 10 hr lifetime: pbl_evolution(nday=3,lifetime=10)
